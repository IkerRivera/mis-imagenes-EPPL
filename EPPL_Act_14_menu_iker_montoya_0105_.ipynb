# redaccion del problema
* Un estudiante de preparatoria necesita calcular diferentes medidas
 geométricas para su clase de matemáticas. Requiere un programa que le
 permita calcular el área y perímetro de un círculo, así como el volumen
 de una esfera, ingresando únicamente el radio de la figura.
 CONTEXTO EDUCATIVO:

* Descripción: El perímetro de un círculo, también conocido como longitud de
la circunferencia, representa la distancia total alrededor del círculo.
Se calcula multiplicando 2 por pi (π) por el radio.

* Descripción: El volumen de una esfera representa la cantidad de espacio
tridimensional que ocupa el cuerpo esférico. Se calcula multiplicando 4/3
por pi (π) por el radio elevado al cubo.

* Descripción: Muestra el menú principal con las opciones disponibles
para el usuario. Presenta una interfaz clara y organizada
con emojis para hacerla más atractiva visualmente.
Parámetros: Ninguno
Retorna: Ninguno

 #  1️⃣ ÁREA DEL CÍRCULO:      A = π * r²
#   2️⃣ PERÍMETRO DEL CÍRCULO:  P = 2 * π * r  
#  3️⃣ VOLUMEN DE LA ESFERA:   V = (4/3) * π * r³
#

```python
# elaborado por Iker Montoya NL = 38

import math

# 🎨 Función para mostrar el menú principal
def menu():
    print("\n" + "="*60)
    print("  MENÚ PRINCIPAL - CÁLCULOS GEOMÉTRICOS 📐")
    print("="*60)
    print(" 🔵 1. Calcular área del círculo")
    print(" 🟢 2. Calcular perímetro del círculo")
    print(" 🔴 3. Calcular volumen de la esfera")
    print(" ⚫ 4. Salir del programa")
    print("="*60)

# 🔵 Función para calcular el área del círculo
def areacirculo():
    print("\n🔵 --- CÁLCULO DEL ÁREA DEL CÍRCULO --- 🔵")
    radio = float(input(" Ingresa el radio del círculo: "))
    area = math.pi * radio ** 2
    print(f"\n ✨ Resultado:")
    print(f" 📏 Radio: {radio}")
    print(f" 📐 Área: {area:.4f} unidades cuadradas")
    print(f" 🎯 Fórmula usada: A = π * r²")

# 🟢 Función para calcular el perímetro del círculo
def perimetrocirculo():
    print("\n🟢 --- CÁLCULO DEL PERÍMETRO DEL CÍRCULO --- 🟢")
    radio = float(input(" Ingresa el radio del círculo: "))
    perimetro = 2 * math.pi * radio
    print(f"\n ✨ Resultado:")
    print(f" 📏 Radio: {radio}")
    print(f" 📐 Perímetro: {perimetro:.4f} unidades")
    print(f" 🎯 Fórmula usada: P = 2 * π * r")

# 🔴 Función para calcular el volumen de la esfera
def volumenesfera():
    print("\n🔴 --- CÁLCULO DEL VOLUMEN DE LA ESFERA --- 🔴")
    radio = float(input(" Ingresa el radio de la esfera: "))
    volumen = (4/3) * math.pi * radio ** 3
    print(f"\n ✨ Resultado:")
    print(f" 📏 Radio: {radio}")
    print(f" 📐 Volumen: {volumen:.4f} unidades cúbicas")
    print(f" 🎯 Fórmula usada: V = (4/3) * π * r³")

# ============================================================================
# 🚀 PROGRAMA PRINCIPAL
# ============================================================================

print("\n" + "🌟"*30)
print(" 🎓 BIENVENIDO AL PROGRAMA GEOMÉTRICO 📐")
print("  Cálculo de Círculos y Esferas")
print("🌟"*30)

opcion = 0

while opcion != 4:
    menu()
    opcion = int(input("\n 💡 Selecciona una opción (1-4): "))

    if opcion == 1:
        areacirculo()
    elif opcion == 2:
        perimetrocirculo()
    elif opcion == 3:
        volumenesfera()
    elif opcion == 4:
        print("\n" + "👋"*30)
        print(" 🎉 programa realizado por iker montoya 0105")
        print(" 📚 ¡Éxito en tus estudios! 🌟")
        print(" 👋"*30 + "\n")
    else:
        print("\n ⚠️  Opción no válida. Por favor elige 1, 2, 3 o 4.")

# Fin del programa
```
